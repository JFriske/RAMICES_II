```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import matplotlib as mpl
import pandas as pd
#from cycler import cycler
#mpl.rcParams['axes.prop_cycle'] = cycler(color='bgrcmyk')
```

```python
kuerzel = 'NucDiskAGBTest'

coldGasMet = pd.read_table('../Output/Test/Enrichment_Absolute_ColdGas.dat',index_col=False , delimiter = ',\s')

coldGasMass = pd.read_table('../Output/Test/Mass.dat',index_col=False, delimiter = ',\s')

coldGasMetLog = pd.read_table('../Output/' + kuerzel + '/Enrichment_Log_ColdGas.dat',index_col=False , delimiter = ',\s')

coldGasMetNuc = pd.read_table('../Output/' + kuerzel + '/Enrichment_Absolute_ColdGas.dat',index_col=False , delimiter = ',\s')

coldGasMassNuc = pd.read_table('../Output/' + kuerzel + '/Mass.dat',index_col=False, delimiter = ',\s')

events = pd.read_table('../Output/' + kuerzel + '/Events.dat',index_col=False, delimiter = ',\s')
```

```python
ringNumber = 50
ringwidth= 0.015/5
```

```python


#coldGasMassNuc0 = coldGasMassNuc[coldGasMassNuc['Radius'] == 0.00375]
gasRings = []
for i in range(ringNumber):
    a = coldGasMassNuc[coldGasMassNuc['Radius'] == (i+ 0.5)*ringwidth ]

    gasRings.append(a)
    
figure(figsize=(8, 6), dpi=150)
for i in range(ringNumber):
    ring = gasRings[i]
    x = ring['Time']
    y = ring['ColdGasMass']
    colors = plt.cm.turbo(np.linspace(0,1,ringNumber))
    plt.plot(x,y, color=colors[i], label = i) 

#plt.legend()
#plt.ylim(0,1e4)
plt.show()

figure(figsize=(8, 6), dpi=150)

for i in range(ringNumber):
    ring = gasRings[i]
    x = ring['Time']
    y = ring['HotGasMass']
    colors = plt.cm.turbo(np.linspace(0,1,ringNumber))
    plt.plot(x,y, color=colors[i], label = i) 

#plt.legend()
plt.show()
```

```python

#coldGasMassNuc0 = coldGasMassNuc[coldGasMassNuc['Radius'] == 0.00375]
gasRings = []
for i in range(ringNumber):
    a = events[events['Radius'] == (i+ 0.5)*ringwidth ]
    gasRings.append(a)
    
    

figure(figsize=(8, 6), dpi=150)

for i in range(ringNumber):
    event = gasRings[i]
    x = event['Time']
    y = event['FormationEfficiency']
    colors = plt.cm.turbo(np.linspace(0,1,ringNumber))
    plt.plot(x,y, color=colors[i], label = i) 
    plt.yscale('log')
#plt.legend()
plt.ylabel('Formation Efficieny')
plt.show()


figure(figsize=(8, 6), dpi=150)

for i in range(ringNumber):
    event = gasRings[i]
    x = event['Time']
    y = event['CCSNRate']
    colors = plt.cm.turbo(np.linspace(0,1,ringNumber))
    plt.plot(x,y, color=colors[i], label = i) 
    plt.yscale('log')
    plt.ylabel('ccSN rate')
    plt.vlines(3.5, -10,1000000)
    
    
figure(figsize=(8, 6), dpi=150)

for i in range(ringNumber):
    event = gasRings[i]
    x = event['Time']
    y = event['AGBRate']
    colors = plt.cm.turbo(np.linspace(0,1,ringNumber))
    plt.plot(x,y, color=colors[i], label = i) 
    plt.yscale('log')
    plt.ylabel('AGB rate')
    plt.vlines(1, -10,100000000)
    plt.vlines(2, -10,100000000)
    plt.vlines(3, -10,100000000)
    plt.vlines(4, -10,100000000)
    plt.vlines(5, -10,100000000)
    plt.vlines(6, -10,100000000)
    plt.vlines(7, -10,100000000)
    
#plt.legend()
plt.show()
```
